### What are embeddings ? 

Embedding is the vector representation of a text , obj, images and audio.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Representation of embedding models

word_embedding={
    "cat":[0.8,0.6],
    "dog":[0.5,0.7],
    "kitten":[0.7,0.8],
    "puppy":[0.65,0.35],
    "car":[0.2,0.4],
    "truck":[0.3,0.5]
}
```

```python
fig,ax = plt.subplots(figsize=(8, 6))

for word, coodinate in word_embedding.items():
    ax.scatter(coodinate[0],coodinate[1],s=100)
    ax.annotate(word,(coodinate[0],coodinate[1]),xytext=(5,5),textcoords="offset points")

ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')

ax.set_title("Simplified word embeddings in 2d space")
ax.grid(True,alpha=0.3)

plt.tight_layout()
plt.show()
```

### Mesuring similarity

```python
def cosine_similarity(vec1,vec2):
    """
    Cosine Similary measures angle between two vectos
    - Result close to 1: Very similar
    - Result close to 0: Not relatable
    - Result close to -1: Opposite
    
    """

    dot_product=np.dot(vec1,vec2)
    norm_a=np.linalg.norm(vec1)
    norm_b=np.linalg.norm(vec2)

    return dot_product/(norm_a * norm_b)


cat_vector=[0.8,0.6,0.5]
kitten_vector=[0.9,0.5,0.5]
dog_vector=[0.5,0.3,0.5]

cat_kitten_similarity=cosine_similarity(cat_vector,kitten_vector)
cat_dog_similarity=cosine_similarity(cat_vector,dog_vector)
print(f"cat_kitten_similarity : {cat_kitten_similarity} and cat_dog_similarity :  {cat_dog_similarity}")
```
