### JSON Parsing

```python
from langchain_community.document_loaders import JSONLoader

json_path = "data/pdf_files/dummy-json.json"

try:
    # Need JqSchema - ie key which one to print
    # text_content - just print string , not json if true
    loader_json = JSONLoader(json_path,".company.employees[]", text_content=False)
    documents = loader_json.load()

    for index,chunk in enumerate(documents):
        print(f"For {index} chunk: content -> {chunk.page_content}")
        print(f"For {index} chunk: meta -> {chunk.metadata}")
except Exception as e:
    print(f"Exception is {e}")
```
