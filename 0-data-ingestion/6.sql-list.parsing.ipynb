## Working with SQL Lite 

```python
import sqlite3

conn=sqlite3.connect('data/data_base/company.db')
cursor=conn.cursor()
```

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS employees
               (id INTEGER PRIMARY KEY,name TEXT, role TEXT,salary REAL)''')
```

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS projects
               (id INTEGER PRIMARY KEY,name TEXT, desciption TEXT,budget REAL)''')
```
